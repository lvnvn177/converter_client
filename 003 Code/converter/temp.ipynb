```python
from langchain_community.document_loaders import PyPDFLoader

path = "KCI_FI002984601.pdf"
loader = PyPDFLoader(path)
pages = loader.load()
```

```python
temp = ''
for page in pages:
    temp += page.page_content
```

```python
print(temp)
```

```python
pages[0].page_content
```

```python
from langchain_ollama import ChatOllama
from langchain_core.prompts import ChatPromptTemplate
```

```python
llm = ChatOllama(
    model="llama3.1:8b",
    temperature=0.6,
    top_p=0.5,
    repeat_penalty=1.5
)
```

```python
prompt = ChatPromptTemplate.from_messages(
        [
            (
                "system",
                "You are a helpful assistant that summarize given text."
                "Answer in Korean",
            ),
            ("human", "{input}"),
        ]
    )
```

```python
chain = prompt | llm
answer = chain.invoke(
    {
        "input": pages[0].page_content,
    }
)
```
